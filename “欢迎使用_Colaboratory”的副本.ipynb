```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
from PIL import Image
import random

# 重新定义加载和解析函数
def load_images():
    building_img = np.array(Image.open("building4.png").convert("L"))
    open_space_img = np.array(Image.open("open space4.png").convert("L"))
    plants_img = np.array(Image.open("plants4.png").convert("L"))
    playground_img = np.array(Image.open("playground4.png").convert("L"))

    height, width = building_img.shape
    campus_grid = np.zeros((height, width), dtype=int)

    campus_grid[building_img > 128] = 1
    campus_grid[open_space_img > 128] = 2
    campus_grid[plants_img > 128] = 3
    campus_grid[playground_img > 128] = 4

    return campus_grid, height, width

# 构造图，减少节点移除概率
def construct_graph(campus_grid, height, width):
    # 创建一个二维网格图
    G = nx.grid_2d_graph(height, width)

    # 遍历网格中的所有节点，应用规则移除不允许的节点
    for i in range(height):
        for j in range(width):
            if campus_grid[i, j] == 1:  # 建筑区域（完全移除，不允许路径通过）
                G.remove_node((i, j))
            elif campus_grid[i, j] == 3:  # 植物区域（部分移除）
                if np.random.rand() < 0.4:  # 降低移除概率为 40%
                    G.remove_node((i, j))
            elif campus_grid[i, j] == 4:  # 操场区域（尽量保留完整性）
                if np.random.rand() < 0.5:  # 降低移除概率为 50%
                    G.remove_node((i, j))
            # 开放空间（campus_grid[i, j] == 2）完全保留，不移除任何节点
            # 其他区域可以根据需要添加规则
    return G

def generate_connected_paths(G, campus_grid, height, width, num_groups=10, paths_per_group=5):
    open_spaces = [(i, j) for i in range(height) for j in range(width) if campus_grid[i, j] == 2 and (i, j) in G]
    playgrounds = [(i, j) for i in range(height) for j in range(width) if campus_grid[i, j] == 4 and (i, j) in G]

    grouped_paths = []

    for group_idx in range(num_groups):
        group = []
        if not open_spaces or not playgrounds:
            print(f"Group {group_idx + 1}: No sufficient points to generate paths.")
            grouped_paths.append([])
            continue  # 如果没有足够的点，跳过

        # 随机选择一个起点
        start = random.choice(open_spaces)
        current_point = start

        try:
            for _ in range(paths_per_group):
                # 从当前点随机选择一个目标点
                end = random.choice(playgrounds + open_spaces)
                if end in G and current_point in G:
                    path = nx.shortest_path(G, source=current_point, target=end)
                    group.append(path)
                    current_point = end  # 更新连接点
        except nx.NetworkXNoPath:
            print(f"Group {group_idx + 1}: Path not found.")
            continue

        grouped_paths.append(group)

    # 确保返回的组数量为 num_groups
    while len(grouped_paths) < num_groups:
        grouped_paths.append([])

    return grouped_paths

# 生成连贯路径（确保有至少 5 个路径组，每组至少 5 条路径）
def generate_connected_paths(G, campus_grid, height, width, num_groups=5, paths_per_group=5):
    open_spaces = [(i, j) for i in range(height) for j in range(width) if campus_grid[i, j] == 2 and (i, j) in G]
    playgrounds = [(i, j) for i in range(height) for j in range(width) if campus_grid[i, j] == 4 and (i, j) in G]

    grouped_paths = []  # 用于存储路径组

    for group_idx in range(num_groups):  # 确保生成指定数量的路径组
        group = []
        attempts = 0  # 重试次数限制
        while len(group) < paths_per_group and attempts < 10:  # 尝试生成完整路径组
            attempts += 1
            if not open_spaces or not playgrounds:
                break  # 如果没有足够的起点或终点，停止生成

            # 随机选择起点和终点
            start = random.choice(open_spaces)
            end = random.choice(playgrounds)

            try:
                # 生成路径并添加到组
                path = nx.shortest_path(G, source=start, target=end)
                group.append(path)

                # 更新当前连接点
                current_point = end
                for _ in range(paths_per_group - 1):  # 尝试生成更多连贯路径
                    next_target = random.choice(playgrounds + open_spaces)
                    if next_target in G and current_point in G:
                        next_path = nx.shortest_path(G, source=current_point, target=next_target)
                        group.append(next_path)
                        current_point = next_target
                        if len(group) >= paths_per_group:
                            break
            except nx.NetworkXNoPath:
                continue  # 如果找不到路径，继续尝试

        grouped_paths.append(group)

    # 确保返回的路径组数量为 num_groups
    while len(grouped_paths) < num_groups:
        grouped_paths.append([])

    return grouped_paths

# 可视化分组路径
def visualize_grouped_paths(campus_grid, grouped_paths, thickness=2):
    for group_idx, paths in enumerate(grouped_paths):
        path_grid = campus_grid.copy()

        if not paths:  # 如果路径组为空，跳过
            print(f"Group {group_idx + 1}: No paths to visualize.")
            continue

        for path in paths:
            for (i, j) in path:
                # 标记路径的中心点
                path_grid[i, j] = 5

                # 增加路径厚度
                for di in range(-thickness, thickness + 1):
                    for dj in range(-thickness, thickness + 1):
                        ni, nj = i + di, j + dj
                        if 0 <= ni < campus_grid.shape[0] and 0 <= nj < campus_grid.shape[1]:
                            path_grid[ni, nj] = 5

        # 显示路径
        plt.figure(figsize=(8, 8))
        plt.imshow(path_grid, cmap="coolwarm")
        plt.title(f"Path Group {group_idx + 1} (Total Paths: {len(paths)})")
        plt.axis("off")
        plt.show()

# 主程序
campus_grid, height, width = load_images()
G = construct_graph(campus_grid, height, width)
grouped_paths = generate_connected_paths(G, campus_grid, height, width, num_groups=5, paths_per_group=5)
visualize_grouped_paths(campus_grid, grouped_paths)


```

```
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# 区域颜色映射
colors = {
    1: "gray",       # 建筑区域
    2: "green",      # 开放空间
    3: "lightgreen", # 植物区域
    4: "blue",       # 操场
    5: "red",        # 路网
}

# 可视化分区并叠加路网
def visualize_with_zones_and_paths(campus_grid, grouped_paths, thickness=2):
    # 创建一个新的图层，用于显示分区
    zone_grid = np.zeros_like(campus_grid)

    # 填充分区
    for i in range(campus_grid.shape[0]):
        for j in range(campus_grid.shape[1]):
            zone_grid[i, j] = campus_grid[i, j]  # 按区域值填充

    # 叠加路径
    for paths in grouped_paths:
        for path in paths:
            for (i, j) in path:
                # 绘制路径到 zone_grid 上
                zone_grid[i, j] = 5  # 标记路径
                for di in range(-thickness, thickness + 1):
                    for dj in range(-thickness, thickness + 1):
                        ni, nj = i + di, j + dj
                        if 0 <= ni < campus_grid.shape[0] and 0 <= nj < campus_grid.shape[1]:
                            zone_grid[ni, nj] = 5

    # 设置颜色映射
    cmap = mcolors.ListedColormap([colors[1], colors[2], colors[3], colors[4], colors[5]])
    bounds = [0.5, 1.5, 2.5, 3.5, 4.5, 5.5]
    norm = mcolors.BoundaryNorm(bounds, cmap.N)

    # 可视化分区和路网
    plt.figure(figsize=(10, 10))
    plt.imshow(zone_grid, cmap=cmap, norm=norm)
    plt.colorbar(ticks=[1, 2, 3, 4, 5], label="Zones and Path")
    plt.title("Campus Zoning with Overlaid Paths")
    plt.axis("off")
    plt.show()

# 主程序
campus_grid, height, width = load_images()
G = construct_graph(campus_grid, height, width)
grouped_paths = generate_connected_paths(G, campus_grid, height, width, num_groups=5, paths_per_group=5)
visualize_with_zones_and_paths(campus_grid, grouped_paths)
```
